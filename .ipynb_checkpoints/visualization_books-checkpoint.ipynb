# Amazon Bestselling Books

Visualization of Amazon's Top 50 bestselling books from 2009 to 2019. 
The dataset has 550 samples of best sellers. What can we learn about the trends of bestsellers?

First we'll import the data and explore while checking for missing values.

```python
# import basic packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
books_df = pd.read_csv('bestsellers with categories.csv') 
```

### What books stay on the bestseller list over the years?

```python
fig = plt.figure(figsize=(8, 8))
ax = sns.stripplot(x='Year', y='Name', data=books_df, order= books_df.Name.value_counts().iloc[:10].index, palette='cool')
ax = sns.boxenplot(x='Year', y='Name', data=books_df, order= books_df.Name.value_counts().iloc[:10].index, palette='icefire_r')
plt.title('Years on Top 10 Bestseller made Top 50')
```

Only one book made the top 50 for 10 years of the 11 years. The Books that stayed bestsellers are Psychology/self-help books and childrens books. Only one classic book made the top 10 bestsellers 5 years and its the only one to have a two year gap and come back. 


### Do fictional or non fictional books rate beter?

```python
fig = plt.figure(figsize=(10, 7))
sns.boxplot(x='Genre', y='User Rating', data= books_df, palette='icefire_r')
sns.swarmplot(x='Genre', y='User Rating', data= books_df, color='.25', size=3, palette='cool')
plt.title('User Ratings for Non Fiction and Fiction Books')
```

Fictional books has a long tail to lower ratings but over all tend to rate higher. Is this a bias of who is reviewing the books?

```python
books_df.Genre.value_counts()
```

```python

fig = plt.figure(figsize=(10, 7))
sns.boxplot(x='Genre', y='Reviews', data= books_df, palette='icefire_r')
sns.swarmplot(x='Genre', y='Reviews', data= books_df, color='.25', size=3, palette='cool')
plt.title('User Reviews for Non Fiction and Fiction Books')
```

```python
sns.countplot(x='Year', data= books_df, hue='Genre')
```

We can tell that even though non fictional books rate lower and have less reviews there are more of them on the best sellers list almost every year. The data seems to have some bias on who is rating and reviewing books. 

### Do Authors with more books have higher ratings?

```python
fig = plt.figure(figsize=(12,10))
ax = sns.countplot(y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:20].index, palette='icefire_r')
plt.title('Authors with the Most Books')
```

```python
fig = plt.figure(figsize=(12,15))
ax = sns.stripplot(x= 'User Rating', y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:25].index, palette='cool', jitter=True)
ax = sns.boxenplot(x= 'User Rating', y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:25].index, palette='icefire_r')
plt.title('Ratings of books by Authors with the most books')
```

Having more books might get more sold but they don't all get good reviews.

### What effect does the Price have on ratings?

```python
fig = plt.figure(figsize=(12,10))
sns.swarmplot(x='User Rating', y='Price', data= books_df, hue='Year', palette='icefire_r',size=3)
plt.title('Ratings vs Price by Year')
```

There is some interesting clustering. All $50 books on the bestseller list rate 4.5 which seems odd. Most of the bestsellers for 2018, and 2019 seem to have particularly high ratings.

### What patters can we find in the number of reviews given?

```python
sns.scatterplot(x='User Rating', y='Reviews', data= books_df, hue='Year', palette='icefire_r', ci=None)
plt.title('Reviews vs Ratings by Year')
```

```python
sns.scatterplot(x='Price', y='Reviews', data= books_df, hue='Year', palette='icefire_r', ci=None)
plt.title('Reviews vs Price by Year')
```

```python
fig = plt.figure(figsize=(12,10))
sns.boxplot(x='Year', y='Reviews', data= books_df, palette='icefire_r')
```

There don't seem to be any corrilations for the number of reviews. 2014 has slightly more reviews than other years.
