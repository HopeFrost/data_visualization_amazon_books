# Amazon Bestselling Books

Visualization of Amazon's Top 50 bestselling books from 2009 to 2019. 
The dataset has 550 samples of best sellers. What can we learn about the trends of bestsellers?

First we'll import the data and explore while checking for missing values.

```python
# import basic packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
books_df = pd.read_csv('bestsellers with categories.csv') 
```

```python
books_df.head()
```

```python
books_df.info()
```

```python
# Great! Data seems pretty clean. Lets make sure theres no duplicats samples
books_df[books_df.duplicated()]
```

```python
books_df.isna().sum()
```

## Lets Investigate:

What information would be useful to a publisher? 
What kinds of books should a small library keep in stock?

* What books stay on the bestseller list over the years?
* Do fictional or non fictional books rate better? 
* Do Authors with more books have higher ratings?
* What effect does the Price have on ratings? 
* What patterns can we find in the number of reviews given?



### What books stay on the bestseller list over the years?

```python
fig = plt.figure(figsize=(8, 12))
ax = sns.stripplot(x='Year', y='Name', data=books_df, order= books_df.Name.value_counts().iloc[:20].index, size=10, palette='icefire_r')
#ax = sns.boxenplot(x='Year', y='Name', data=books_df, order= books_df.Name.value_counts().iloc[:20].index, palette='icefire_r')
plt.title('Years on Top 20 Bestseller made Top 50')
```

```python
books_df.Name.value_counts().iloc[:50].index
```

Only one book made the top 50 for 10 years of the 11 years. The Books that stayed bestsellers are Psychology/self-help books and children's books. 18 books made top 50 at least 5 years. 


### Do fictional or non fictional books rate better?

```python
fig = plt.figure(figsize=(10, 7))
sns.boxplot(x='Genre', y='User Rating', data= books_df, palette='icefire_r')
sns.swarmplot(x='Genre', y='User Rating', data= books_df, color='.25', size=3, palette='cool')
plt.title('User Ratings for Non Fiction and Fiction Books')
```

Fictional books has a long tail to lower ratings but over all tend to rate higher. Is this a bias of who is reviewing the books?

```python
books_df.Genre.value_counts()
```

```python

fig = plt.figure(figsize=(10, 7))
sns.boxplot(x='Genre', y='Reviews', data= books_df, palette='icefire_r')
sns.swarmplot(x='Genre', y='Reviews', data= books_df, color='.25', size=3, palette='cool')
plt.title('User Reviews for Non Fiction and Fiction Books')
```

```python
fig = plt.figure(figsize=(10, 7))
sns.boxplot(x='Genre', y='Price', data= books_df, palette='icefire_r')
sns.swarmplot(x='Genre', y='Price', data= books_df, color='.25', size=3, palette='cool')
plt.title('price for Non Fiction and Fiction Books')
```

```python
sns.countplot(x='Year', data= books_df, hue='Genre', palette='icefire')
plt.title('Fiction vs Non Finction Bestsellers Each Year')
```

We can tell that even though non fictional books rate lower and have less reviews there are more of them on the best sellers list almost every year. The data seems to have some bias on who is rating and reviewing books. 

### Do Authors with more books have higher ratings?

```python
fig = plt.figure(figsize=(12,10))
ax = sns.countplot(y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:20].index, palette='icefire_r')
plt.title('Top 20 Authors with the Most Books')
```

```python
fig = plt.figure(figsize=(12,20))
ax = sns.swarmplot(x= 'User Rating', y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:25].index, size=4, palette='cool')
ax = sns.boxenplot(x= 'User Rating', y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:25].index, palette='icefire_r')
plt.title('Ratings of books by Authors with the most books')
```

```python
fig = plt.figure(figsize=(12,8))
ax = sns.swarmplot(x= 'User Rating', y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:10].index, size=3, palette='cool')
ax = sns.boxenplot(x= 'User Rating', y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:10].index, palette='icefire_r')
plt.title('Top 10 Ratings of books by Authors with the most books')
```

```python
fig = plt.figure(figsize=(12,11))
ax = sns.swarmplot(x= 'Price', y='Author', data=books_df, order= books_df.Author.value_counts().iloc[:25].index, size=3, palette='cool')
```

```python
fig = plt.figure(figsize=(12,15))
ax = sns.swarmplot(x= 'Price', y='Name', data=books_df, order= books_df.Name.value_counts().iloc[:25].index, size=3, palette='cool')
```

Having more books might get more sold but they don't all get good reviews.

### What effect does the Price have on ratings?

```python
fig = plt.figure(figsize=(14,10))
sns.swarmplot(x='User Rating', y='Price', data= books_df, hue='Year', palette='icefire_r',size=3)
plt.title('Ratings vs Price by Year')
```

There is some interesting clustering. All $50 books on the bestseller list rate 4.5 which seems odd (probably academic books). Most of the bestsellers for 2018, and 2019 seem to have particularly high ratings and lower price tags.

### What patterns can we find in the number of reviews given?

```python
sns.scatterplot(x='User Rating', y='Reviews', data= books_df, hue='Year', palette='icefire_r', ci=None)
plt.title('Reviews vs Ratings by Year')
```

```python
sns.scatterplot(x='Price', y='Reviews', data= books_df, hue='Year', palette='icefire_r', ci=None)
plt.title('Reviews vs Price by Year')
```

```python
fig = plt.figure(figsize=(8,7))
sns.boxplot(x='Year', y='Reviews', data= books_df, palette='icefire_r')
```

There don't seem to be any correlations of reviews. 2014 has slightly higher reviews than other years.



### Publishers Gleanings:
* People seem to be most interested in Psychology or Self Help books and Children's Books.  
* While Fictional books get more reviews and higher ratings more people are buying Non Fictional books
* There seems to be a trend in the last couple years for bestsellers to have high ratings and low prices. This may indicate a competitive market.


### Libraries interest:
* Should have largest sections dedicated to Psychology, Self Help and Children's Books.
* Well known Authors are read more even if the specific book gets a lower ratings.
* Should focus on lower price range books maybe some academic books from the upper price range.
